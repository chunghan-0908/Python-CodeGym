# Function

## Function(hàm) là gì?

Hàm (function) là một khối lệnh đặc biệt, nó được đặt tên, giúp mã chương trình dễ đọc hơn, và có thể gọi để sử dụng ở các nơi khác nhau trong chương trình. Hàm chính là khối lệnh có thể tái sử dụng (reuseable).

### Hàm

Cú pháp:

`def <tên hàm>(<tham số>):
    //Nội dung hàm`

Gọi hàm:
`<tên hàm>(<giá trị đầu vào>)`

### Ví dụ: Hàm in

```python
# Viết 1 hàm để giới thiệu về khóa học 
def print_course_info():
    print("******************************")
    print("*Course: Python Fundamental  *")
    print("*Lesson 13                   *")
    print("*Organizationn:CodeGym Online*")
    print("******************************")
```

```python
print_course_info()
```

```python
# Hàm tính thể tích hình lập phương


#side (parameter): -> biến số
#giá trị được truyền vào: -> đối số
def calc_cube_volume(side):
    print(f"Cude volume is:  {side ** 3}")
    
calc_cube_volume(5)
calc_cube_volume(10)
calc_cube_volume(15)
```

```python
#Hàm xếp loại học sinh
#Tính điểm tb 3 môn học toán, văn, anh
#Cho biết học lực giỏi >= 6.5 / khá (5 - > 6.5) / trung bình ( <5)
def calc_avg(math, literature,  english):
    average = (math + literature + english)/3
    
    if average >= 6.5:
        print(average)
        print("Học sinh giỏi")
    elif 5 <= average < 6.5:
        print(average)
        print("Học sinh khá")
    else:
        print(average)
        print("Học sinh trung bình")
        
calc_avg(8,10,9)
```

### Thực hành: Viết hàm nhận vào một số nguyên và kiểm tra xem số đó có phải là số nguyên tố hay không.


```python
def is_prime_number(number):
    #Giả định (assumption) number là số nguyên tố
    is_prime = True #giả định number là số nguyên tố
    for i in range(2, number):
        if number % i == 0:
            print(f"{number} is not a prime number")
            is_prime = False #Cập nhật lại number k phải là SNT
            break #dừng lại từ lần đầu tiên tìm dc 
    if is_prime:
        print(f"{number} is a prime number")

is_prime_number(49)
```

# return

Cú pháp:

`return <giá trị trả về>`

##### Lưu ý: Sau khi thực hiện lệnh return, chương trình sẽ thoát khỏi hàm ngay lập tức (những lệnh sau đó không được thực hiện)

Ví dụ:

Hãy viết 1 hàm trả về giá trị trung bình của 2 số bất kỳ

```python
def calc_average (num1, num2):
    return ((num1+num2)/2)
    
result = calc_average(4,5)
print(result)
```

```python
# Hàm tính chiều dài của chuỗi
# Viết hàm nhận vào một chuỗi và trả về chiều dài của chuỗi
def calc_length_string(string):
    return len(string)

result = calc_length_string("Chung Hân")
print(result)
```

```python
# Viết hàm nhận vào một số nguyên và trả về giai thừa của số nguyên đó
def calc_factorial(number):
    factorial = 1
    for i in range (2, number + 1):
        factorial *= i
    return factorial

print(calc_factorial(5))
```
