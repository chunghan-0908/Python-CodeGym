## Bài 1: reversed list
Hãy viết một chương trình để đảo ngược các phần tử trong list

```python
aList = [100, 200, 300, 400, 500]
```

**Output:**

```python
[500, 400, 300, 200, 100]
```

```python
aList = [100, 200, 300, 400, 500]
# code below this line
aList.reverse()
print(aList)
# code above this line
```

```python
aList = [100, 200, 300, 400, 500]
print(aList[::-1])
```

```python
aList = [100, 200, 300, 400, 500]

reverse_list = []
for i in range(len(aList) -1, -1, -1):
    reverse_list.append(aList[i])
    
print(reverse_list)
```

## Bài 4: concatenation 2
Tạo ra một list mới từ 2 list ban đầu theo thứ tự như sau:

```python
list1 = ["Hello ", "take "]
list2 = ["Dear", "Sir"]
```

**Output**

```python
['Hello Dear', 'Hello Sir', 'take Dear', 'take Sir']
```

```python
list1 = ["Hello ", "take "]
list2 = ["Dear", "Sir"]
# code below this line
list3 = []
for i in range(len(list1)):
    for j in range(len(list2)):
        n = list1[i] + list2[j]
        list3.append(n)
print(list3)
# code above this line
```

## Bài 7: list replacement

```python
list1 = [5, 10, 15, 20, 25, 50, 20]
```

**Output**

## Bài 8: list removement
Hãy xóa bỏ **tất cả** số 20 ra khỏi list

```python
list1 = [5, 20, 15, 20, 25, 50, 20]
```

**Output:**

```python
[5, 15, 25, 50]
```

```python
list1 = [5, 20, 15, 20, 25, 50, 20]
# code below this line
for i in list1:
    if i == 20:
        list1.remove(20)
print(list1)
# code above this line
```

## Bài 12: is it the same?
Hãy kiểm tra xem các phần tử trong tuple có giống nhau hay không.

```python
tuple1 = (45, 45, 45, 45)
```

**Output**

```python
True
```

```python
tuple1 = (45, 45, 45, 45)
# code below this line
n = tuple1[0]
for i in tuple1:
    if i != n:
        print(False)
        break
else:
    print(True)
# code above this line
```
