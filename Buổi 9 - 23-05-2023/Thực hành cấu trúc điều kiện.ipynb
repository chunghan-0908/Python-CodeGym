### Bài 16: Viết một chương trình thực hiện phép tính đơn giản như sau:
```
    Nhập số thứ nhất: 16
    Nhập số thứ hai: 9
    Nhập toán tử (+, -, *, /): +
    Kết quả: 16 + 9 = 25
```

````python
so1 = float(input("Nhập số thứ nhất: "))
so2 = float(input("Nhập số thứ hai: "))
toantu = input("Nhập toán tử (+,-,*,/): ")
if toantu == "+":
    kq = so1 + so2
    print("Kết quả: ", kq)
elif toantu == "-":
    kq = so1 - so2
    print("Kết quả: ", kq)
elif toantu == "*":
    kq = so1 * so2
    print("Kết quả: ", kq)
elif toantu == "/":
    kq = so1 / so2
    print("Kết quả: ", kq)
else:
    print("Không tìm thấy toán tử")
````
