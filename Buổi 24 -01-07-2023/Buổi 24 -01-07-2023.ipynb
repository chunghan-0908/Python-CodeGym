1.	Xây dựng một ứng dụng tkinter để quản lý nhà sách dựa theo ứng dụng tkinter mẫu đã học
2.	Thông tin của mỗi quyển sách cần quản lý: tên sách, thể loại và giá tiền
3.	Các chức năng cần có: thêm sách mới, tìm kiếm và hiển thị thông tin sách, xóa sách, cập nhật giá tiền của sách
4.	Các widget cần sử dụng: Entry, Label, Button, MessageBox. Có thể sử dụng thêm các widget khác như combobox, TreeView

```python
class Sach:
    def __init__(self, ten, theloai, giatien):
        self.ten = ten
        self.theloai = theloai
        self.giatien = giatien
    
    def __str__(self):
        return f"Tên sách: {self.ten}, thể loại: {self.theloai}, giá {self.giatien}"
```

```python
from tkinter import *
from tkinter import messagebox

window = Tk()
window.title("Hệ thống quản lý sách")

sach_list = []

#Tạo label
lbl_title = Label(window, text="HỆ THỐNG QUẢN LÝ SÁCH")
lbl_ten = Label(window, text="Tên sách: ")
lbl_theloai = Label(window, text="Thế loại: ")
lbl_giatien = Label(window, text="Giá tiền: ")
lbl_timkiem = Label(window, text="Tìm kiếm: ")

#Tạo entry
ten = Entry(window)
theloai = Entry(window)
giatien = Entry(window)
timkiem = Entry(window)

#Tạo button
btn_them = Button(window, text="THÊM", command=them_sach)
btn_capnhat = Button(window, text="CẬP NHẬT", command=capnhat_sach)
btn_xoa = Button(window, text="XÓA", command=xoa_sach)
btn_timkiem = Button(window, text="TÌM KIẾM", command=timkiem_sach)

#Tính năng thêm
def them_sach():
    ten_value = ten.get() #Lấy giá trị mà user đã nhập vào Entry ten
    theloai_value = theloai.get()
    giatien_value = giatien.get()
    
    if ten_value == "" or theloai_value == "" or giatien_value =="":
        messagebox.showerror("Sach", "Hãy nhập tên, thể loại hoặc giá tiền!")
    else:
        sachmoi = Sach(ten_value, theloai_value, float(giatien_value))
        sach_list.append(sachmoi)
        
        print([sach.ten for sach in sach_list]) #In list sách
        messagebox.showinfo("Sach", f"Thêm sách {ten_value} thành công!")
        
#Tính năng xóa
def xoa_sach():
    sach_can_xoa = timkiem.get()
    check = False 
    
    for i in range(len(sach_list)):
        if sach_list[i].ten ==  sach_can_xoa:
            del sach_list[i]
            print([sach.ten for sach in sach_list])
            messagebox.showinfo("XoaSach",f"Xóa {sach_can_xoa} thành công!")
            check = True
            break
        if check == False:
            messagebox.showerror("XoaSach", f"Không tìm thấy sách {sach_can_xoa}")

#Tính năng tìm kiếm
def timkiem_sach():
    timsach = timkiem.get()
    check = False
    
    for i in range(len(sach_list)):
        if sach_list[i].ten ==  timsach:
            print([sach.ten for sach in sach_list])
            messagebox.showinfo("TimSach",sach_list[i])
            check = True
            break
        if check == False:
            messagebox.showerror("TimSach", f"Không tìm thấy sách {timsach}") 
            
#Tính năng cập nhật 
def capnhat_sach():
    sach_can_cap_nhat = ten.get()
    check = False
    
    for i in range(len(sach_list)):
        if sach_list[i].ten == sach_can_cap_nhat:
            sach_list[i].theloai = theloai.get()
            sach_list[i].giatien = float(giatien.get())
            print([sach.ten for sach in sach_list])
            messagebox.showinfo("CapNhatSach",f"Cập nhật thông tin sách {sach_can_cap_nhat} thành công!")
            check = True
            break
        if check == False:
            messagebox.showerror("CapNhatSach", f"Không tìm thấy sách {sach_can_cap_nhat}") 
            
#Gắn tọa độ 
lbl_title.grid(row=0, column=2)
lbl_ten.grid(row=1, column=1)
lbl_theloai.grid(row=2, column=1)
lbl_giatien.grid(row=3, column=1)
lbl_timkiem.grid(row=1, column=3)

ten.grid(row=1, column=2)
theloai.grid(row=2, column=2)
giatien.grid(row=3, column=2)
timkiem.grid(row=1, column=4)

btn_them.grid(row=4, column=1)
btn_capnhat.grid(row=4, column=2)
btn_xoa.grid(row=4, column=3)
btn_timkiem.grid(row=2, column=4)

window.mainloop()
```
