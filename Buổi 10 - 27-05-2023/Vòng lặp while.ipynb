## Cấu trúc lặp while

### Ví dụ: In ra tất cả số chẵn từ 0 đến n

```python
n = int(input("Nhập vào n: "))
i = 0 #Khởi tạo biến đếm
while i <= n:
    print(i)
    i += 2
```

### Ví dụ: In ra tất cả số lẻ từ n về 1

```python
#Vd n = 10
n = int(input("Nhập n: "))
i = 0 #tạo biến đếm
while i <= n:
    if n % 2 != 0:
        print(n)
        i -= 2
```

```python
n = int(input("Nhập n: "))
i = 0
while i >= n:
    if i % 2 == 1:
        print(i)
        i -= 1
```

### Ví dụ: Đảo ngược chuỗi

```python
content = input("Nhập chuỗi: ")
index = len(content) - 1
while index >= 0:
    print(content[index], end="")
    index -= 1
```

### Ví dụ: Tam giác vuông

```python
chieucao = int(input("Nhập vào chiều cao tam giác: "))
i = 1
while i <= chieucao:
    print("*" * i)
    i += 1
```

### Lệnh Break

Cú pháp:

`break`

Lệnh break sẽ giúp hủy vòng lặp đang thực hiện

##### Lưu ý:

##### - Lệnh break <b>BẮT BUỘC</b> phải được đặt trong vòng lặp
##### - Lệnh break sẽ hủy vòng lặp gần nhất (trong trường hợp có nhiều vòng lắp lồng vào nhau)

#### Ví dụ

Sử dụng lệnh `while (True)` và break và in ra 10 lần câu "Hello mọi người"

```python
lottery_data = "0001010001000111011"
i = 0
count = 0
while i <= len(lottery_data):
    if lottery_data[i] == "1":
        count += 1
        print(f"Người thứ {i} trúng giải")
        if count == 3:
            break
    i += 1
```

## Lệnh continue

```python
#Viết chương trình in ra tất cả các số nguyên chia hết cho 3 và 7 khoảng 100 đến 200
i = 100
n = 200
while i <= n:
    if i % 3 == 0 and i % 7 == 0:
        print(i)
    i += 1
```

```python
#Viết chương trình in ra tất cả các số nguyên chia hết cho 3 và 7 khoảng 100 đến 200 trừ 168 
i = 100
n = 200
while i <= n:
    if i % 3 == 0 and i % 7 == 0 and i != 168:
        print(i)
    i += 1
```

```python
#Khác
i = 100
n = 200
while i <= n:
    i += 1
    if i == 168 or i == 147:
        continue
    if i % 3 == 0 and i % 7 == 0:
        print(i)
    i += 1
```
